## Changing the column Price into integer.

```python
df["Price"]
```

```python
Price_Updated = []
for i in df["Price"]:
    i = i.replace("₹","")
    if "Cr" in i:
        i = i.replace("Cr", "* 10000000")
    if "Lac" in i:
        i = i.replace("Lac", "* 100000")
    i = i.split("*")
    float_i_0 = float(i[0])
    int_i_1 = int(i[-1])
    mul = round(float_i_0 * int_i_1)
    Price_Updated.append(mul)
df["Price_Updated"] = Price_Updated    
```

```python
df.head()
```

## Removing "Beds" from the column.

```python
list1 = []
for a in df["beds"]:
    a = a.replace("Beds","")
    list1.append(a)
df["beds"] = list1
```

```python
Carpet_Area = []
for i in df["Carpet Area"]:
    if isinstance(i, float):  # Check if i is a float
        Carpet_Area.append(i)  # Append the float value directly
    else:
        if "sqm" in i:
            i = i.replace("sqm", "")
        elif "," in i:
            i = i.replace(",", "")
        elif "sqyrd" in i:
            i = i.replace("sqyrd", "")
            i = float(i)
            i = round(i * 0.83612736)  # To convert sqyrds to sqms
        elif " Sq-ft  2 " in i:
            i = i.replace(" Sq-ft  2 ", "")
            i = float(i)
            i = round(i * 0.092903)
        Carpet_Area.append(i)
df["Carpet Area"] = Carpet_Area
```

```python
BHK_2 = []
df["BHK"] = df["BHK"].astype(str)
for j in df["BHK"]:
    j = j.strip()
    if len(j) > 1:
        j = j[-1]
    BHK_2.append(j)   
```
