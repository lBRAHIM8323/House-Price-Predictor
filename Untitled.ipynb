```python
Price_Updated = []
for i in df["Price"]:
    i = i.replace("₹","")
    if "Cr" in i:
        i = i.replace("Cr", "* 10000000")
    if "Lac" in i:
        i = i.replace("Lac", "* 100000")
    i = i.split("*")
    float_i_0 = float(i[0])
    int_i_1 = int(i[-1])
    mul = round(float_i_0 * int_i_1)
    Price_Updated.append(mul)
df["Price_Updated"] = Price_Updated    
```
